```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_baseline = pd.read_json("../models/logs/logistic_baseline_tf_loss.json")
df_baseline_tfidf = pd.read_json("../models/logs/logistic_baseline_tfidf_loss.json")
```

```python
def loss_curve_plot(df, file_name):
    plt.plot(df["current_iter"], df["train_loss"])
    plt.xlabel("current_iter")
    plt.ylabel("train_loss")
    plt.title("Training Loss Curve")

    plt.savefig(f"../figures/loss_curve_{file_name}.png", dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()
```

```python
loss_curve_plot(df_baseline, "baseline")
```
